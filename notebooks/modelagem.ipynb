## Informações sobre o Dataset
1. Attribute Information:
   -- Only 14 used
      -- 1. #3  (age):      age in years
      -- 2. #4  (sex):      (1 = male; 0 = female)
      -- 3. #9  (cp):       chest pain type
        -- Value 1: typical angina
        -- Value 2: atypical angina
        -- Value 3: non-anginal pain
        -- Value 4: asymptomatic
      -- 4. #10 (trestbps):  resting blood pressure (in mm Hg on admission to the 
        hospital)
      -- 5. #12 (chol):   serum cholestoral in mg/dl   
      -- 6. #16 (fbs):    (fasting blood sugar > 120 mg/dl)  (1 = true; 0 = false)   
      -- 7. #19 (restecg): esting electrocardiographic results
        -- Value 0: normal
        -- Value 1: having ST-T wave abnormality (T wave inversions and/or ST 
                    elevation or depression of > 0.05 mV)
        -- Value 2: showing probable or definite left ventricular hypertrophy
                    by Estes' criteria   
      -- 8. #32 (thalach): maximum heart rate achieved
      -- 9. #38 (exang):   exercise induced angina (1 = yes; 0 = no)  
      -- 10. #40 (oldpeak): ST depression induced by exercise relative to rest   
      -- 11. #41 (slope):   the slope of the peak exercise ST segment
        -- Value 1: upsloping
        -- Value 2: flat
        -- Value 3: downsloping     
      -- 12. #44 (ca):     number of major vessels (0-3) colored by flourosopy        
      -- 13. #51 (thal):     thal: 3 = normal; 6 = fixed defect; 7 = reversable defect 
      
      -- 14. #58 (num)       (the predicted attribute): diagnosis of heart disease (angiographic disease status)
        -- Value 0: < 50% diameter narrowing
        -- Value 1: > 50% diameter narrowing
        (in any major vessel: attributes 59 through 68 are vessels)

```python
def load_data(data_path,data_file):
    local_path = os.path.join(data_path,data_file)
    names = ['idade','sexo','dor','pressao','colesterol','acucar','exame','batimentos','exercicio','depressao','declive',
            'vasos', 'tal','diagnostico']
    return pd.read_csv(local_path, header=None,names=names)
```

```python
heart_data = load_data(HEART_PATH, HEART_FILE)
```

# Passo 4: KMeans

```python
heart_nao_rotulado = heart_data.drop('diagnostico', axis=1)
```

```python
modelo = KMeans(n_clusters=5)
modelo.fit(heart_nao_rotulado)
grupos = modelo.predict(heart_nao_rotulado)
```

```python
print(grupos)
```

## 4.1: Visualização

```python
xs = heart_data.iloc[:,0]
ys = heart_data.iloc[:,2]

centroids = modelo.cluster_centers_
centroids_x = centroids[:,0]
centroids_y = centroids[:,2]
```

```python
_= plt.scatter(xs, ys, c=grupos, alpha=.5)
_= plt.scatter(centroids_x, centroids_y, c='black', marker='X', s=100)
```

## 4.2: Avaliação

```python
df = pd.DataFrame({'grupos':grupos, 'diagnostico':heart_data['diagnostico']})
ct = pd.crosstab(df['grupos'],df['diagnostico'])
print(ct)
```
